```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data to plot
controls = [203, 203.05, 202.65, 193.65]
treatment = [210.8, 214.28, 213.59, 213.75]

# Create a DataFrame for the boxplot
data = pd.DataFrame({
    'Bleach': controls,
    'ML-1': treatment
})

# Create boxplot
fig, ax = plt.subplots(figsize=(3, 4))
data.boxplot(ax=ax, showfliers=False, capprops=dict(linewidth=0))
ax.set_ylabel('Break stress', fontsize=12)

# Increase tick label font sizes
ax.tick_params(axis='both', which='major', labelsize=10)

# Set y-axis limits from 200 to 250
ax.set_ylim(190, 255)

# Add individual data points with jitter
np.random.seed(123)  # For consistent jitter positioning
jitter_strength = 0.1

# Add points for Control group (x=1)
x_control = np.ones(len(controls)) + np.random.uniform(-jitter_strength, jitter_strength, len(controls))
ax.scatter(x_control, controls, alpha=0.6, color='blue', s=20, zorder=3)

# Add points for Treatment group (x=2)
x_treatment = 2 * np.ones(len(treatment)) + np.random.uniform(-jitter_strength, jitter_strength, len(treatment))
ax.scatter(x_treatment, treatment, alpha=0.6, color='orange', s=20, zorder=3)

# Add horizontal line at y = 248 labeled "Virgin"
ax.axhline(y=248, color='red', linestyle='-', linewidth=1, label='Virgin')
ax.legend(fontsize=10, loc='upper right')

# Hide right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Remove the grid
ax.grid(False)

plt.tight_layout()
plt.show()
```

```python
# Calculate median of control values (Bleach)
median_control = np.median(controls)
virgin_value = 248

# Calculate percent decrease from virgin to median control
percent_decrease = ((virgin_value - median_control) / virgin_value) * 100

print(f"Virgin value: {virgin_value}")
print(f"Median of control values: {median_control}")
print(f"Percent decrease from virgin to median control: {percent_decrease:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data to plot
controls = [203, 203.05, 202.65, 193.65, 198.39, 195.64]
treatment = [210.8, 214.28, 213.59, 213.75]
olaplex = [194.71, 198]

# Create DataFrame with treatment (ML-1) in the last position
data = pd.concat([
    pd.Series(controls, name='Bleach'),
    pd.Series(olaplex, name='Olaplex No. 3'),
    pd.Series(treatment, name='ML-1'),
], axis=1)

# Rest of your code remains the same
fig, ax = plt.subplots(figsize=(3.5, 4))
data.boxplot(ax=ax, showfliers=False, capprops=dict(linewidth=0))
ax.set_ylabel('Break stress', fontsize=12)
ax.tick_params(axis='both', which='major', labelsize=10)
ax.set_ylim(190, 255)

# Calculate medians for the gray area
bleach_median = np.median(controls)
treatment_median = np.median(treatment)

# Add light gray area between medians
ax.axhspan(bleach_median, treatment_median, alpha=0.2, color='gray', zorder=1)

# Add individual data points with jitter
np.random.seed(123)
jitter_strength = 0.1

# Bleach group (x=1)
x_control = np.ones(len(controls)) + np.random.uniform(-jitter_strength, jitter_strength, len(controls))
ax.scatter(x_control, controls, alpha=0.6, color='blue', s=20, zorder=3)

# Olaplex group (x=2)
x_olaplex = 2 * np.ones(len(olaplex)) + np.random.uniform(-jitter_strength, jitter_strength, len(olaplex))
ax.scatter(x_olaplex, olaplex, alpha=0.6, color='green', s=20, zorder=3)

# ML-1 treatment group (x=3, now last)
x_treatment = 3 * np.ones(len(treatment)) + np.random.uniform(-jitter_strength, jitter_strength, len(treatment))
ax.scatter(x_treatment, treatment, alpha=0.6, color='orange', s=20, zorder=3)

ax.axhline(y=248, color='red', linestyle='-', linewidth=1, label='Virgin')
ax.legend(fontsize=10, loc='upper right')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.grid(False)

plt.tight_layout()
plt.show()
```
